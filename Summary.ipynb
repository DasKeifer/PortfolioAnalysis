```python
%matplotlib inline

import pickle
import matplotlib.pyplot as plt
```

```python
def showResultsSummary(results, filterStrings=[""], divsor=1, ylim=[0,10]):
    print("{:12}, {:10}, {:10}, {:10}, {:10}, {:10}, {:10}".format("  id", "  avg", "  max", "  min", "ttl std", "std avg", "shp avg"))
    for result in results:
        name = result["params"][0][0]['ticker'] + str(result["params"][0][0]['targetPercent'])
        
        found = False
        for filt in filterStrings:
            if filt in name:
                found = True
                break
        if(not found):
            continue
            
        if len(result["params"][0]) > 1:
            name += result["params"][0][1]['ticker']
            
        print("{:12}, {:10.4f}, {:10.4f}, {:10.4f}, {:10.4f}, {:10.4f}, {:10.4f}".format(
            name, result["total_avg"]/divsor, result["total_max"]/divsor, result["total_min"]/divsor, result["total_stddev"]/divsor, result["dailyStddev_avg"], result["sharpe_avg"]))
        ax = (result["totals"]/divsor).rename(name).plot(legend=True, figsize=(16,8)) 
        ax.set_ylim(ylim)
```

```python
with open(r'data\analysis\starting100k_5yr_1992.pickle', 'rb') as f:
    res_5_100k = pickle.load(f)
with open(r'data\analysis\starting100kPeriodic_5yr_1992.pickle', 'rb') as f:
    res_5_100kPer = pickle.load(f)
with open(r'data\analysis\startingOnlyPeriodic_5yr_1992.pickle', 'rb') as f:
    res_5_OnlyPer = pickle.load(f)
```

```python
showResultsSummary(res_5_100k, ["spy","0.6"], 100000, [0,12])
```

```python
showResultsSummary(res_5_100kPer, ["spy","0.6"], 100000 + 1600 * 5 * 365.25/14, [0,6])
```

```python
showResultsSummary(res_5_OnlyPer, ["spy","0.6"], 1600 * 5 * 365.25/14, [0,6])
```
